```python
import pandas as pd


# Load the Excel file
file_path = 'Champlain Capital Partners IV LP Q1 2024 Form 468.pdf.xlsx'
xls = pd.ExcelFile(file_path)

# Initialize the dictionary with the same keys as your original script
data = {

    "Portfolio Company Name": [],
    "Employer ID": [],
    "Critical Technology": [],
    "Financing Type": [],
    "Investment Type": [],
    "Financing Description": [],
    "Initial Financing Date": [],
    "Equity Capital Investment": [],
    "Interest or Dividend Rate": [],
    "Ownership % (Fully Diluted)": [],
    "Restructured": [],
    "Loan / Debt Status": [],
    "Total Cash Invested (A)": [],
    "Cost at Beginning Period": [],
    "Loan": [],
    "Debt Equity": [],
    "Portfolio Securities Recievables from Sale of": [],
    "Assets Acquired": [],
    "Assets Acquired Operating Concerns Acquired": [],
    "Notes and Other Equity in Reinvestor/Relender": [],
    "Exception": [],
    "Totals": [],

}
def extract_value(df, header):
    for col in df.columns:
        if df[col].str.contains(header, na=False).any():
            value = df[df[col].str.contains(header, na=False)].iloc[0, df.columns.get_loc(col) + 1]
            return value
    return None

# Iterate through each sheet in the Excel file and populate the dictionary
for sheet_name in xls.sheet_names:
    df = pd.read_excel(xls, sheet_name)

    # Example: assuming the first row is the header and the rest are the values
    for column in df.columns:
        if column in data:
            data[column].extend(df[column].dropna().tolist())

# Print the populated dictionary
for key, value in data.items():
    print(f"{key}: {value}")
```

```python
import numpy as np
def merge_rows(group):
    if group.empty:
        return pd.Series(dtype='float64')

    merged_row = group.iloc[0].copy()
    for col in group.columns:
        if col not in ['Financing Type']:
            merged_row[col] = group[col].dropna().iloc[0] if not group[col].dropna().empty else np.nan
    merged_row['Total Cash Invested (A)'] = group['Total Cash Invested (A)'].sum()
    merged_row['Cost at Beginning Period'] = group['Cost at Beginning Period'].sum()
    return merged_row

# Group by 'Financing Type' and apply the merging function
result = df.groupby('Financing Type').apply(merge_rows).reset_index(drop=True)

result = result.dropna(subset=['Portfolio Company Name'])

# Convert 'Interest or Dividend Rate' to percentages
result['Interest or Dividend Rate'] = result['Interest or Dividend Rate'].apply(lambda x: f"{x*100:.0f}%" if pd.notnull(x) else x)
result['Ownership\n% (Fully Diluted)'] = result['Ownership\n% (Fully Diluted)'].apply(lambda x: f"{x*100:.0f}%" if pd.notnull(x) and isinstance(x, (int, float)) else x)
# Format 'Total Cash Invested (A)' and 'Cost at Beginning Period' with dollar signs and commas
result['Total Cash Invested (A)'] = result['Total Cash Invested (A)'].apply(lambda x: f"${x:,.0f}" if pd.notnull(x) else x)
result['Cost at Beginning Period'] = result['Cost at Beginning Period'].apply(lambda x: f"${x:,.0f}" if pd.notnull(x) else x)

result = result.drop(result.index[0])

```

```python
def standardize_equity_type(value):
    if isinstance(value, str):  # Check if the value is a string
        if 'common' in value.lower():
            return 'Common'
        elif 'preferred' in value.lower():
            return 'Preferred'
    return value

# Apply the function to the column
final['Type of Equity Security'] = final['Type of Equity Security'].apply(standardize_equity_type)
final.rename(columns={'Equity Pricing - Equity Ownership (FD)': ' Equity Ownership (FD)'}, inplace=True)
final["# of Full time Employees"]=fte
final = final.head(2)
```
